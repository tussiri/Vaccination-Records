```python
def parse_vaccine_data(text):
    country_vaccine_data = {}
    countries = re.split(r'(?=\n[A-Z](?:[A-Z ])*(?=\n))', text)

    for country in countries:
        lines = country.strip().split('\n')
        country_name = lines[0].strip()

        if "see" in country_name.lower():
            country_vaccine_data[country_name] = {
                "See": country_name.split(" see ")[1]}
            continue

        vaccine_info = ' '.join(lines[1:])
        vaccine_info = re.sub(r'\n', ' ', vaccine_info)

        yellow_fever_entry = re.search(
            r"Country requirement at entry:(.*?)(?=WHO)", vaccine_info)
        yellow_fever_recommendation = re.search(
            r"WHO vaccination recommendation:(.*?)(?=Malaria|$)", vaccine_info)
        malaria_risk = re.search(r"Malaria risk(.*?)(?=WHO)", vaccine_info)
        malaria_prevention = re.search(
            r"WHO recommended prevention in risk areas:(.*?)(?=Other)", vaccine_info)
        other_requirements = re.search(
            r"Other country requirement\(s\)(.*?)$", vaccine_info)

        country_data = {
            "Yellow fever": {
                "Country requirement at entry": "null",
                "WHO vaccination recommendation": "null"
            },
            "Malaria": {
                "Risk": "null",
                "WHO recommended prevention in risk areas": "null"
            },
            "Other requirements": "null"
        }

        if yellow_fever_entry:
            country_data["Yellow fever"]["Country requirement at entry"] = yellow_fever_entry.group(
                1).strip()
        if yellow_fever_recommendation:
            country_data["Yellow fever"]["WHO vaccination recommendation"] = yellow_fever_recommendation.group(
                1).strip()
        if malaria_risk:
            country_data["Malaria"]["Risk"] = malaria_risk.group(1).strip()
        if malaria_prevention:
            country_data["Malaria"]["WHO recommended prevention in risk areas"] = malaria_prevention.group(
                1).strip()
        if other_requirements:
            country_data["Other requirements"] = other_requirements.group(
                1).strip()

        country_vaccine_data[country_name] = country_data

    return country_vaccine_data


with open("/Users/tumainiussiri/Desktop/Web-Dev-Projects/Projects/Vaccination-Records/vaccination-record-backend/assets/vaccination-requirements-and-who-recommendations-ith-2022-country-list.txt", "r") as file:
    text = file.read()

     # Parse the vaccine data
    country_vaccine_data = parse_vaccine_data(text)
```

```python
import json


def save_to_json_file(data, filename):
    with open(filename, 'w') as file:
        json.dump(data, file, indent=2)


save_to_json_file(country_vaccine_data, 'country_vaccine_data.json')
```
